```python
def max_crossing_sum(arr, l, m, r):
    # Include elements on left of mid
    left_sum = float('-inf')
    total = 0
    for i in range(m, l - 1, -1):
        total += arr[i]
        if total > left_sum:
            left_sum = total

    # Include elements on right of mid
    right_sum = float('-inf')
    total = 0
    for i in range(m + 1, r + 1):
        total += arr[i]
        if total > right_sum:
            right_sum = total

    # Return sum of elements on left and right of mid
    return left_sum + right_sum


def max_subarray_sum(arr, l, r):
    # Base case
    if l == r:
        return arr[l]

    # Find middle point
    m = (l + r) // 2

    # Return maximum of three cases
    return max(
        max_subarray_sum(arr, l, m),
        max_subarray_sum(arr, m + 1, r),
        max_crossing_sum(arr, l, m, r)
    )


# Main driver code
T = int(input())
for _ in range(T):
    N = int(input())
    A = list(map(int, input().split()))
    print(max_subarray_sum(A, 0, N - 1))
```
