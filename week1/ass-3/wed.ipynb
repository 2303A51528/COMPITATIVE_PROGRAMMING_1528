```python
def max_crossing_sum(arr, l, m, r):

    left_sum = float('-inf')
    total = 0
    for i in range(m, l - 1, -1):
        total += arr[i]
        if total > left_sum:
            left_sum = total

    right_sum = float('-inf')
    total = 0
    for i in range(m + 1, r + 1):
        total += arr[i]
        if total > right_sum:
            right_sum = total

    return left_sum + right_sum


def max_subarray_sum(arr, l, r):

    if l == r:
        return arr[l]

    m = (l + r) // 2
    return max(
        max_subarray_sum(arr, l, m),
        max_subarray_sum(arr, m + 1, r),
        max_crossing_sum(arr, l, m, r)
    )


T = int(input())
for _ in range(T):
    N = int(input())
    A = list(map(int, input().split()))
    print(max_subarray_sum(A, 0, N - 1))
```
