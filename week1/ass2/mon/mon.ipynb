```python
def max_meetings(meetings):
    # Sort meetings by end time
    meetings.sort(key=lambda x: x[1])

    count = 0
    last_end = -1

    for s, e in meetings:

        if s >= last_end:
            count += 1
            last_end = e

    return count


T = int(input())
for _ in range(T):
    N = int(input())
    meetings = []
    for _ in range(N):
        s, e = map(int, input().split())
        meetings.append((s, e))

    print(max_meetings(meetings))

```
